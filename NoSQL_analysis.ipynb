### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": {"$eq":20}}

# Use count_documents to display the number of documents in the result
print("Number of establishments with hygiene score of 20: ", 
      establishments.count_documents(query), "\n")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene20 = list(establishments.find(query))
hygiene20_df = pd.DataFrame(hygiene20)

# Display the number of rows in the DataFrame
print(f"The number of rows in this dataframe are {len(hygiene20_df)}\n")

# Display the first 10 rows of the DataFrame
hygiene20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

This requires execution of NoSQL_setup.ipynb prior as the setup file converted the RatingValue to an Integer.

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName":{"$regex":"London"}, "RatingValue":{"$gte":4}}

# Use count_documents to display the number of documents in the result
print("Number of London Local Authority establishments with Rating Value >= 4: ", 
      establishments.count_documents(query), "\n")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

This requires execution of NoSQL_setup.ipynb prior as the setup file converted the RatingValue to an Integer and latitude, longitude to doubles.

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Hygiene score must be present (not None)
# Excluding Penang Flavours

degree_search = 2

pf_latitude_dict = establishments.find_one({"BusinessName":"Penang Flavours"}, {"geocode.latitude":1})
latitude = pf_latitude_dict["geocode"]["latitude"]

pf_longitude_dict = establishments.find_one({"BusinessName":"Penang Flavours"}, {"geocode.longitude":1})
longitude = pf_longitude_dict["geocode"]["longitude"]

# https://www.mongodb.com/docs/manual/reference/operator/query/geoWithin/#mongodb-query-op.-geoWithin
# https://stackoverflow.com/questions/25734092/query-locations-within-a-radius-in-mongodb
# https://www.mongodb.com/docs/manual/reference/operator/query/centerSphere/#mongodb-query-op.-centerSphere

query = {"RatingValue":{"$eq":5},
         "geocode":{"$geoWithin":{"$centerSphere":[[longitude, latitude], degree_search]}},
         "scores.Hygiene":{"$ne":None},
         "BusinessName":{"$ne":"Penang Flavours"}
        }

# Optionally reduce fields for debug activity
# fields = {"AddressLine1":1, "AddressLine2":1, "AddressLine3":1, "AddressLine4":1, 
#           "BusinessName":1, "BusinessType": 1, "LocalAuthorityName":1, 
#           "RatingDate":1, "RatingKey":1, "RatingValue":1, 
#           "geocode":1, "scores":1}

# Sort by hygiene score
sort =  [("scores.Hygiene",1)]

# Get top 5 only
limit = 5

# Store the query in a variable
top5_by_PF = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
print(f"Top results within 0.01 degrees of Penang Flavours (long {longitude}, lat {latitude})")
pprint(top5_by_PF)
```

```python
# Convert result to Pandas DataFrame
print(f"Top results within 0.01 degrees of Penang Flavours (long {longitude}, lat {latitude})")
top5_by_PF_df = pd.DataFrame(top5_by_PF)
top5_by_PF_df
```
